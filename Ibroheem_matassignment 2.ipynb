```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
fig, axes = plt.subplots()

axes.plot(x, y, color = 'green',label = 'X square', lw = 2, alpha = .9, marker = '*', markersize = 9, markeredgecolor = 'red', markerfacecolor = 'yellow' )

axes.legend(loc = 0)
```

```python
fig, axes = plt.subplots(figsize = (10.3, 5.3))

axes.plot(y+10, y, color = 'blue', label = 'y+10', lw = 2, marker = '*',markersize = 9, markerfacecolor = 'purple', markeredgewidth = .9 , markeredgecolor = 'yellow' )
axes.plot(y**2, y, color = 'red', label = 'y**2', alpha = 1.3,lw = 2, ls = '--', marker = 'o', markersize = 8, markerfacecolor = 'yellow', markeredgewidth = 1, markeredgecolor = 'green')
axes.plot(y*10, y, color = 'purple',alpha = .8, label = 'y+3', lw = 3, marker = 'o', markersize = 8, markerfacecolor = 'red', markeredgewidth = 1, markeredgecolor = 'green', drawstyle = 'steps')

axes.legend(loc = 0)
```
