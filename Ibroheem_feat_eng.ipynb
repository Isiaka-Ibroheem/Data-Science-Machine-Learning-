Cleaning BsmtFin SF 1, BsmtFin SF 2, Bsmt Unf SF and Total Bsmt SF

*  Owning to the fact that only one value is missing in the three columns below , i chose to fill it up with their mean. and the total basement in square feet is the sum of of basement 1, basement 2 and unfinished basement so filling up the other three automatically fill up the total

```python
#cleaning BsmtFin SF 1, BsmtFin SF 2, Bsmt Unf SF and Total Bsmt SF  

houses['BsmtFin SF 1'].fillna(houses['BsmtFin SF 1'].mean(), inplace = True)
houses['BsmtFin SF 2'].fillna(houses['BsmtFin SF 2'].mean(), inplace = True)
houses['Bsmt Unf SF'].fillna(houses['Bsmt Unf SF'].mean(), inplace = True)
houses['Total Bsmt SF'] = houses['BsmtFin SF 1'] + houses['BsmtFin SF 2'] + houses['Bsmt Unf SF']
houses.head()
```
