```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(5, 100, 20)
y = np.linspace(20, 150, 20)
```

### Step 1: Create a figure with 6 subplots (3 rows) with the basic matplotlib command plt

- plot 1:  x-axis = x , y-axis = y  . Green color
- plot 2:  x-axis = y , y-axis = x  . blue color
- plot 3:  x-axis = x^2 , y-axis = y  . yellow color
- plot 4:  x-axis = x  , y-axis =  y^2 . red color
- plot 5:  x-axis = x+2 , y-axis =  y+10 . magenta color
- plot 6:  x-axis = y*10 , y-axis =  x . cyan color

**label and title each plot appropriately**

**width of 8 inches, height of 12 inches**

```python
plt.figure(figsize = (8, 12))

#plot1
plt.subplot(3, 2, 1)
plt.plot(x, y, color = 'g')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of Y against X')
plt.tight_layout()

#plot2
plt.subplot(3, 2, 2)
plt.plot(y, x, color = 'b')
plt.xlabel('Y')
plt.ylabel('X')
plt.title('Plot of X against Y')
plt.tight_layout()

#plot3
plt.subplot(3, 2, 3)
plt.plot(x**2, y, color = 'y')
plt.xlabel('X sqr')
plt.ylabel('Y')
plt.title('Plot of Y against X^2')
plt.tight_layout()

#plot4
plt.subplot(3, 2, 4)
plt.plot(x, y**2, color = 'r')
plt.xlabel('X')
plt.ylabel('Y sqr')
plt.title('Plot of Y^2 against X')
plt.tight_layout()

#plot5
plt.subplot(3, 2, 5)
plt.plot(x+2, y+10, color = 'm')
plt.xlabel('X + 2')
plt.ylabel('Y + 10')
plt.title('Plot of Y+10 against X+2')
plt.tight_layout()

#plot6
plt.subplot(3, 2, 6)
plt.plot(y+10, x, color = 'c')
plt.xlabel('Y + 10')
plt.ylabel('X')
plt.title('Plot of X against Y+10')
```

### Step 2: Create a figure with 2 subplots using the object orriented method with three line plots each

- plot 1
    - line 1:  x+2, x
    - line 2:  x+10, x
    - line 3:  x, x^2
    
- plot 2
    - line 1:  y/2, y
    - line 2:  y+10, y
    - line 3:  y, y^(1/2)
    
    
**show the legends**

**label each plot appropriately**

**width = 8, height = 4; dpi=100**

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (8, 4), dpi = 100)
plt.tight_layout()

axes[-2].plot(x+2, x, 'b', label = 'x against x+2')
axes[-2].plot(x+10, x, 'r', label = 'x against x+10')
axes[-2].plot(x, x**2, 'g', label = 'x^2 against x')

axes[-2].set_xlabel('X')
axes[-2].set_ylabel('Y')
axes[-2].set_title('Plot 1')
axes[-2].legend(  loc = 0)

axes[-1].plot(y/2, y, 'k', label ='y vs y/2')
axes[-1].plot(y+10, y, 'c', label = 'y vs y+10')
axes[-1].plot(y, y**0.5, 'm', label = 'y^.5 vs y ')

axes[-1].set_xlabel('X')
axes[-1].set_ylabel('Y')
axes[-1].set_title('Plot 2')
plt.tight_layout()
axes[-1].legend(loc = 0)

```
