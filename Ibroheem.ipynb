In a class of 10 students (Kemi, Dapo, Eze, Chike, David, Wale, Hammid, Ruth, Ada, Haliya). They had 5 tests, and they had the following scores respectively

- **Mathematics** : 22, 26, 19, 25, 16, 21, 23, 28, 24, 17
- **Chemistry** : 29, 26, 24, 25, 22, 21, 19, 18, 27, 20
- **English** : 23, 35, 27, 26, 31, 34, 28, 22, 39, 21
- **Physics**: 22, 11, 39, 32, 30, 14, 23, 17, 19, 10


```python
#Create a Dataframe with the following columns:
###################
#- Name
#- Maths
#- Chm
#- Eng
#- Phy
maths_scores = [ 22, 26, 19, 25, 16, 21, 23, 28, 24, 17]
chem_scores = [ 29, 26, 24, 25, 22, 21, 19, 18, 27, 20]
eng_scores = [23, 35, 27, 26, 31, 34, 28, 22, 39, 21]
phy_scores = [22, 11, 39, 32, 30, 14, 23, 17, 19, 10]

test_data = pd.DataFrame({'Name': ['Kemi', 'Dapo', 'Eze', 'Chike', 'David', 'Wale', 'Hammid', 'Ruth', 'Ada', 'Haliya'], 'Maths': maths_scores, 'Chem': chem_scores, 'Eng': eng_scores, 'Phy': phy_scores})
test_data
```

6

```python
#create a new column called 'Total_score' containing each student's Total score

test_data['Total_score'] = test_data['Maths'] + test_data['Chem'] + test_data['Eng'] + test_data['Phy']
test_data
```

7

```python
#create a new column called 'Average_score' containing each student's average score
test_data['Average_score'] = test_data['Total_score'] / 4

test_data
```

8

```python
#drop the Total_score column
test_data.drop('Total_score', axis = 1, inplace = True)

test_data
```
