### HANDLING THE MISSING VALUES IN THE DATASET 

```python
#DROP COLUMNS WITH MORE THAN 5% NULL VALUES
drop_percent = df.isnull().sum() / df.shape[0]*100
drop = drop_percent[drop_percent > 5]
df = df.drop(drop.index, axis = 1)
```

```python
#DROP TEXT COLUMNS WITH ONE OR MORE THAN MISSING VALUE
text = df.select_dtypes(include = 'object')
text_drop = text.isnull().sum()
drop = text_drop[text_drop >=1]
df = df.drop(drop.index, axis = 1)
```

```python
df.shape
```

```python
df.isnull().sum() / df.shape[0]
```

```python
df['BsmtFin SF 1'].fillna(value = 0, inplace = True)
df['BsmtFin SF 2'].fillna(value = 0, inplace = True)
df['Bsmt Unf SF'].fillna(value = 0, inplace = True)
df['Total Bsmt SF'] = df['BsmtFin SF 1'] + df['BsmtFin SF 2'] + df['Bsmt Unf SF']
```

```python
df['Garage Cars'].fillna(value = 0.0, inplace = True)
df['Garage Area'].fillna(value = 0.0, inplace = True)
```

```python
df['Mas Vnr Area'].fillna(df['Mas Vnr Area'].mean(), inplace = True)
```

```python
df['Bsmt Full Bath'].fillna(df['Bsmt Full Bath'].mode()[0], inplace = True)
df['Bsmt Half Bath'].fillna(df['Bsmt Half Bath'].mode()[0], inplace = True)
```
